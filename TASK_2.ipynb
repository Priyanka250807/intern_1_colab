```python
#data visualization
import matplotlib.pyplot as plt
# Aggregate the data by 'Sex'
sex_counts = df['Sex'].value_counts()
# Create a bar plot
plt.figure(figsize=(6, 4))  # Increase figure size
plt.bar(sex_counts.index, sex_counts.values, color=['blue', 'pink'])
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

```

```python
import matplotlib.pyplot as plt
# Convert 'Fare' column to numeric, replacing "no value" with NaN
df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')

# Plotting the relationship between Age and Fare
plt.figure(figsize=(10, 6))
plt.plot(df['Age'], df['Fare'], marker='o', linestyle='-', color='b')
plt.title('Relationship between Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame

# Convert 'Fare' column to numeric, handling errors
df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')

# Optional: Aggregate by 'Ticket' if it has many duplicates
ticket_fare = df.groupby('Ticket')['Fare'].mean().reset_index()

# Encode 'Ticket' as numeric indices
ticket_fare['Ticket_Encoded'] = range(len(ticket_fare))

# Create the scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(ticket_fare['Ticket_Encoded'], ticket_fare['Fare'], marker='o', color='b')
plt.title('Average Fare for Each Ticket')
plt.xlabel('Ticket (Encoded)')
plt.ylabel('Average Fare')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(8, 6))
plt.boxplot(df['Age'].dropna(), vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.show()
```

```python
# Pie Chart
if 'Sex' in df.columns:
    sex_counts = df['Sex'].value_counts()
    plt.figure(figsize=(8, 8))
    plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'pink'])
    plt.title('Gender Distribution')
    plt.show()

```

```python
# Histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Fare'].dropna(), bins=20, color='purple', edgecolor='black')
plt.title('Histogram of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Interpolation Chart
if 'Age' in df.columns:
    df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
    df_sorted = df.sort_values(by='Age').dropna(subset=['Age', 'Fare'])
    plt.figure(figsize=(10, 6))
    plt.plot(df_sorted['Age'], df_sorted['Fare'], linestyle='-', marker='o', color='green')
    plt.title('Interpolated Relationship between Age and Fare')
    plt.xlabel('Age')
    plt.ylabel('Fare')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
