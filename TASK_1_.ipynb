```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: Population of India (in billions) from 2000 to 2023
years = np.arange(2000, 2024)
population = [1.05, 1.07, 1.09, 1.11, 1.13, 1.15, 1.17, 1.19, 1.21, 1.23,
              1.25, 1.27, 1.29, 1.31, 1.33, 1.35, 1.37, 1.39, 1.41, 1.42,
              1.43, 1.44, 1.45, 1.46]

# Plot configuration
fig, ax = plt.subplots(2, 1, figsize=(10, 8))

# Bar chart
ax[0].bar(years, population, color='blue')
ax[0].set_title("India's Population Over the Years (2000-2023)")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("Population (in Billions)")
ax[0].grid(True, linestyle='--', alpha=0.6)

# Histogram
ax[1].hist(population, bins=5, color='green', edgecolor='black')
ax[1].set_title("Distribution of India's Population (2000-2023)")
ax[1].set_xlabel("Population (in Billions)")
ax[1].set_ylabel("Frequency")
ax[1].grid(True, linestyle='--', alpha=0.6)

# Adjust layout
plt.tight_layout()
plt.show()
```
